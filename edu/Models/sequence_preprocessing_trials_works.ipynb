```python
text = "I love my family"
import sklearn as sk
from sklearn.feature_extraction.text import CountVectorizer, normalize
```

```python
import labeled_functions
X_corpus , y = labeled_functions.load_with_path()
# Now we have entire corpus with words.

from sklearn.model_selection import train_test_split
X_train_corpus , X_test_corpus, y_train, y_test = \
    train_test_split(X_corpus, y, test_size=1./3, random_state=42)
# Train and Test data are split, still all words.


token = r"(?u)\b[\w\'/]+\b"
vectorizer = CountVectorizer(token_pattern=token, 
                             min_df=5, 
                             stop_words=["the","a","of","and","br","to"],
                             ngram_range=(1,3))
# Now, we have a matrix of ngrams that occurs.

X_train_vector = vectorizer.fit_transform(X_train_corpus)
vectorizer.get_feature_names()
vectorizer.vocabulary_
X_test_vector = vectorizer.transform(X_test_corpus)
```

```python
# print(X_test_vector.toarray()[0])
def convert_to_sequence(vector):
    ''' Takes a vector and converts it into a sequence.
    Args:
        vector: An output of CountVectorizer. i.e X_train_vector, X_test_vector
    '''
    document_sequence = []

    for k in vector.toarray():
        sequence = []
        [sequence.append(i) for i,j in enumerate(k) if j]
        document_sequence.append(sequence)

    return document_sequence

# convert_to_sequence(X_train_vector)[0]
```
